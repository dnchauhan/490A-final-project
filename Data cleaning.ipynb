```python
import pandas as pd
import numpy as np

df = pd.read_csv("yelp_reviews.csv")
df.drop(df.columns[0],axis=1,inplace=True)
df
```

```python
# The counts of each star rating

# 5    17254
# 4     8299
# 3     2948
# 1     1387
# 2     1369
```

```python
# df = df.groupby(['Animal'])

df_5 = df[df['stars']==5]
df_4 = df[df['stars']==4]
df_2 = df[df['stars']==2]
df_1 = df[df['stars']==1]

df_neut_raw = df[df['stars']==3]
df_pos_raw = pd.concat([df_4, df_5])
df_neg_raw = pd.concat([df_1, df_2])
```

```python
df_neut = df_neut_raw.sample(frac=1)[:2750]
df_neg = df_neg_raw.sample(frac=1)[:2750]
df_pos = df_pos_raw.sample(frac=1)[:2750]

df_neut['label'] = 'neutral'
df_neg['label'] = 'negative'
df_pos['label'] = 'positive'

res = pd.concat([df_neut, df_neg, df_pos]).sample(frac=1)
```

```python
# export to csv
res.to_csv("restaurant_reviews.csv")
```
